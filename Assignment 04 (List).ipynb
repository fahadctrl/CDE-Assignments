1. Write a program that accepts a list from user and print the alternate element of list.

```python
n = int(input("Enter number of elements: "))
my_list = []
for i in range(n):
    element = input(f"Enter element {i+1}: ")
    my_list.append(element)

print("Alternate elements:")
for i in range(0, len(my_list), 2):
    print(my_list[i])
```

2. Write a program that accepts a list from user. Your program should reverse the content of list and display it. Do not use reverse() method.

```python
n = int(input("Enter number of elements: "))
my_list = []
for i in range(n):
    element = input(f"Enter element {i+1}: ")
    my_list.append(element)

reversed_list = []
for i in range(len(my_list)-1, -1, -1):
    reversed_list.append(my_list[i])

print("Reversed list:", reversed_list)
```

3. Find and display the largest number of a list without using built-in function max(). Your program should ask the user to input values in list from keyboard.

```python
n = int(input("Enter number of elements: "))
numbers = []
for i in range(n):
    num = int(input(f"Enter number {i+1}: "))
    numbers.append(num)

largest = numbers[0]
for num in numbers:
    if num > largest:
        largest = num

print("Largest number:", largest)
```

4. Write a program that rotates the element of a list so that the element at the first index moves to the second index, the element in the second index moves to the third index, etc., and the element in the last index moves to the first index.

```python
n = int(input("Enter number of elements: "))
my_list = []
for i in range(n):
    element = input(f"Enter element {i+1}: ")
    my_list.append(element)

if len(my_list) > 0:
    last_element = my_list.pop()
    my_list.insert(0, last_element)

print("Rotated list:", my_list)
```

5. Write a program that input a string and ask user to delete a given word from a string.

```python
text = input("Enter a string: ")
word_to_delete = input("Enter word to delete: ")

result = ""
current_word = ""
i = 0

while i < len(text):
    if text[i] == ' ':
        if current_word != word_to_delete:
            result += current_word + " "
        current_word = ""
    else:
        current_word += text[i]
    i += 1

if current_word != word_to_delete:
    result += current_word

print("String after deletion:", result.strip())
```

6. Write a program that reads a string from the user containing a date in the form mm/dd/yyyy. It should print the date in the form March 12, 2021.

```python
date_str = input("Enter date (mm/dd/yyyy): ")
month_num, day, year = date_str.split('/')

months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]

month_name = months[int(month_num) - 1]
print(f"{month_name} {int(day)}, {year}")
```

7. Write a program with a function that accepts a string from keyboard and create a new string after converting character of each word capitalized.

```python
def capitalize_words(sentence):
    result = ""
    capitalize_next = True
    
    for char in sentence:
        if capitalize_next and char.isalpha():
            result += char.upper()
            capitalize_next = False
        else:
            result += char
            if char == ' ':
                capitalize_next = True
    return result

text = input("Enter a string: ")
print("Capitalized:", capitalize_words(text))
```

8. Find the sum of each row of matrix of size m x n.

```python
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

matrix = []
for i in range(rows):
    row = []
    for j in range(cols):
        element = int(input(f"Enter element [{i}][{j}]: "))
        row.append(element)
    matrix.append(row)

for i in range(rows):
    row_sum = 0
    for j in range(cols):
        row_sum += matrix[i][j]
    print(f"Sum of row {i+1} = {row_sum}")
```

9. Write a program to add two matrices of size n x m.

```python
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

print("Enter first matrix:")
matrix1 = []
for i in range(rows):
    row = []
    for j in range(cols):
        element = int(input(f"Enter element [{i}][{j}]: "))
        row.append(element)
    matrix1.append(row)

print("Enter second matrix:")
matrix2 = []
for i in range(rows):
    row = []
    for j in range(cols):
        element = int(input(f"Enter element [{i}][{j}]: "))
        row.append(element)
    matrix2.append(row)

result = []
for i in range(rows):
    row = []
    for j in range(cols):
        row.append(matrix1[i][j] + matrix2[i][j])
    result.append(row)

print("Sum of matrices:")
for row in result:
    print(row)
```

10. Write a program to multiply two matrices

```python
r1 = int(input("Enter rows for first matrix: "))
c1 = int(input("Enter columns for first matrix: "))

print("Enter first matrix:")
matrix1 = []
for i in range(r1):
    row = []
    for j in range(c1):
        element = int(input(f"Enter element [{i}][{j}]: "))
        row.append(element)
    matrix1.append(row)

r2 = int(input("Enter rows for second matrix: "))
c2 = int(input("Enter columns for second matrix: "))

if c1 != r2:
    print("Matrix multiplication not possible!")
else:
    print("Enter second matrix:")
    matrix2 = []
    for i in range(r2):
        row = []
        for j in range(c2):
            element = int(input(f"Enter element [{i}][{j}]: "))
            row.append(element)
        matrix2.append(row)

    result = []
    for i in range(r1):
        row = []
        for j in range(c2):
            sum_val = 0
            for k in range(c1):
                sum_val += matrix1[i][k] * matrix2[k][j]
            row.append(sum_val)
        result.append(row)

    print("Product of matrices:")
    for row in result:
        print(row)
```
