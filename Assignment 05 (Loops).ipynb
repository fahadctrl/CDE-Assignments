2. Write a Python program to print the numbers from 20 to 1 using a while loop.

```python
i = 20
while i >= 1:
    print(i)
    i -= 1
```

7. Write a program that takes a number n as input from the user and generates the first n terms of the series formed by squaring the natural numbers.

```python
n = int(input("Enter a number: "))
print(f"The first {n} terms of the series are:")
for i in range(1, n + 1):
    print(i * i, end=" ")
```

9. Write a Python program to print the first 8 terms of an arithmetic progression starting with 3 and having a common difference of 4.

```python
start = 3
difference = 4
for i in range(8):
    term = start + i * difference
    print(term, end=" ")
```

10. Write a Python program to print the first 6 terms of a geometric sequence starting with 2 and having a common ratio of 3.

```python
start = 2
ratio = 3
for i in range(6):
    term = start * (ratio ** i)
    print(term, end=" ")
```

11. Write a program that asks the user for a positive integer value and calculates the sum of all the integers from 1 up to the number entered.

```python
n = int(input("Enter a positive integer: "))
total = 0
for i in range(1, n + 1):
    total += i
print(f"The sum of numbers from 1 to {n} is: {total}")
```

12. Write a program that takes a positive integer N as input and calculates the sum of the reciprocals of all numbers from 1 up to N.

```python
n = int(input("Enter a positive integer: "))
total = 0
for i in range(1, n + 1):
    total += 1 / i
print(f"The sum of reciprocals from 1 to {n} is: {total:.2f}")
```

13. Write a program that prompts the user to enter a number and repeats this process 5 times, then displays the final running total.

```python
total = 0
for i in range(5):
    num = int(input("Enter a number: "))
    total += num
print(f"The final running total is: {total}")
```

14. Write a program that prompts the user to enter a positive integer and calculates its factorial.

```python
n = int(input("Enter a positive integer: "))
if n < 0:
    print("Factorial does not exist for negative numbers")
elif n == 0:
    print("Factorial of 0 is 1")
else:
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    print(f"Factorial of {n} is {factorial}")
```

15. Write a Python program that prompts the user to enter a base number and an exponent, and calculates the power without using ** or math.pow().

```python
base = float(input("Enter base number: "))
exponent = int(input("Enter exponent: "))

if exponent >= 0:
    result = 1
    for i in range(exponent):
        result *= base
    print(f"{base} raised to the power {exponent} is: {result}")
else:
    result = 1
    for i in range(-exponent):
        result *= base
    print(f"{base} raised to the power {exponent} is: {1/result}")
```
